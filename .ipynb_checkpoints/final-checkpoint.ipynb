<h1> Importing Basic Libraries </h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
import random
import pandas as pd
import seaborn as sns
import tensorflow as tf
import tensorflow.keras as keras
import cv2
from tqdm import tqdm
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import image
import os
import glob
import warnings
warnings.filterwarnings("ignore") 
```

```python
#checking if gpu is available on this device
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU'))) 
```

<h1> Reading the Dataset </h1>

```python
#the driving dataset contains images in jpg format of various frames captured while driving
#the jpg images are named as- 0.jpg, 1.jpg,.....45405.jpg

#there is a text file named data which has the steering wheel angles in degrees for images from 0.jpg to 45405.jpg

xs = [] #will store names of all the images
ys = [] #will store the corresponding steering wheel angles

with open("driving_dataset/data.txt") as f:
    for line in f:
        xs.append("driving_dataset/" + line.split()[0])
        #the steering wheel angles are converted from degrees to radians as a method of normalization
        ys.append(float(line.split()[1]) * scipy.pi / 180) 
```

```python
#checking if performed above steps properly
print(xs[100:110]) #will print the name of 10 images 
print(ys[100:110]) #will print corresponding steering wheel angles in radians
```

<h1> Data Preprocessing </h1>

```python
#here the train and validation data is preprocessed to be used by the neural network

#reading an image using opencv
image = cv2.imread('driving_dataset/images/2600.jpg') #--imread() helps in loading an image into jupyter including its pixel values
#displaying the image
#cv2.imshow("sample",image)
# cv2.waitKey(5000) 
# cv2.destroyAllWindows()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
print("image shape: ",image.shape)
```

```python
#understanding the shape of the image
height, width, channels=image.shape
print("height  of an image in pixels: ",height)
print("width  of an image in pixels: ",width)
print("number of channels of an image: ",channels)
```

```python
#we need to reshape the image in accordance with the architecture of the neural network
image = cv2.resize(image, dsize=(200, 66)) #1st arg=new width,2nd arg=new height

#plotting the resized image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
#loading all images and doing the preprocessing

data=[]

for i in tqdm(range(len(xs))):
    path=xs[i].split('/')
    path=path[0]+'/images/'+path[1] #defining the image path
    image = cv2.imread (path)
    image=cv2.resize(image, dsize=(200, 66)) #1st arg=new width,2nd arg=new height
    image=image/255.0 #normalizing pixel values between 0 and 1
    data.append (image)
```

<h1> Splitting the Data- Train, Validation, Test </h1>

```python
#train=90 percent
#test=10 percent---will remain untouched until the training ends

train_xs = data[0:int(len(xs) * 0.9)]
train_ys = ys[0:int(len(xs) * 0.9)]

test_xs = data[int(len(xs) * 0.9):]
test_ys = ys[int(len(xs) * 0.9):]
```

```python
#train is further shuffled (to break the sequence) and splitted into train and validation (85-15 split)

xy_train=list(zip(train_xs,train_ys)) #zipped
random.shuffle(xy_train)
x,y=zip(*xy_train) #unzipping data to get shuffled x and y

#training data final
train_xs=x[0:int(len(x)*0.85)]
train_ys=y[0:int(len(y)*0.85)]

#validation data final
validation_xs=x[int(len(x)*0.85):]
validation_ys=y[int(len(x)*0.85):]
```

```python
num_train=len(train_xs)
num_val=len(validation_xs)
num_test=len(test_xs)

print("number of training images: ",num_train)
print("number of validation images: ",num_val)
print("number of test images: ",num_test)
```

```python
#zipping data

xy_train=list(zip(train_xs,train_ys))
xy_val=list(zip(validation_xs,validation_ys))
xy_test=list(zip(test_xs,test_ys))
```

```python
#final data-

#train- train_xs, train_ys, zipped- xy_train
#val-validation_xs, validation_ys, zipped-xy_val
#test-test_xs, test_vs, zipped-xy_test
```

```python
#the above data is too large to be fed as input at once
#so we define data generators which sens in batches and in proper format- note the x are lists of image vectors
#cnn takes numpy arrays only
```

<h1> Defining the Data Generators </h1>

```python
#https://www.youtube.com/watch?v=EkzB6PJIcCA&list=PLd9i_xMMzZF6mxod8D1CfqhMy85dyLMq2&index=8

def data_generator(data, batch_size):
    """
    Yields the next training batch
    
    """
    #data-list of tuples:[(x1,y1),(x2,y2),(x3,y3).....]
    num_samples = len(data)
    while True: # Loop forever so the generator never terminates
        
        #shuffling and unzipping data before each epoch
        #shuffling-You want to shuffle your data after each epoch because 
        #you will always have the risk to create batches that are not representative of the overall dataset,
        #and therefore, your estimate of the gradient will be off. 
        #Shuffling your data after each epoch ensures that you will not be "stuck" with too many bad batches
        
        random.shuffle(data) #shuffling data before each epoch
        x,y=zip(*data) #unzipping data to get shuffled x and y
        
        # Get index to start each batch: [0, batch_size, 2*batch_size, ..., max multiple of batch_size <= num_samples]
        for offset in range(0, num_samples, batch_size): #completion of loop defines one epoch
            # Get the samples you'll use in this batch
            x_batch_sample = x[offset:offset+batch_size]
            y_batch_sample = y[offset:offset+batch_size]

            # Make sure they're numpy arrays (as opposed to lists)-cnn don't take lists as input
            x_batch = np.array(x_batch_sample)
            y_batch = np.array(y_batch_sample)

            # The generator-y part: yield the next training batch            
            yield x_batch,y_batch
```

```python
#creating train and validation generator

batch_size=64

train_generator=data_generator(xy_train,batch_size)

validation_generator=data_generator(xy_val,batch_size)
```

<h1> Defining the CNN Architecture </h1>

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, Dropout, Lambda, BatchNormalization
from keras.initializers import TruncatedNormal, Constant
from tensorflow.keras import regularizers

#create model
model = Sequential()
#convolutional layers
model.add(Conv2D(24,kernel_size=5,strides=(2,2),activation='relu', kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=TruncatedNormal(stddev=0.1),
                 kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),input_shape=(66,200,3)))
model.add(BatchNormalization())
model.add(Conv2D(36, kernel_size=5, strides=(2,2),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
model.add(Conv2D(48, kernel_size=5, strides=(2,2),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=3, strides=(1,1),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=3, strides=(1,1),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
#flattening layers
model.add(Flatten())
#dense layers with dropout
model.add(Dense(1164,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(100,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(50,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(10,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=TruncatedNormal(stddev=0.1)))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(1,activation='tanh',kernel_initializer=TruncatedNormal(stddev=0.1), 
                 bias_initializer=TruncatedNormal(stddev=0.1)))
```

<h1> Summary of the CNN </h1>

```python
model.summary()
```

<h1> Compiling the Model </h1>

```python
#learning rate- 1e-4 (as mentioned in the paper)
#loss- mean square error (regression problem)

model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-4),
              loss=keras.losses.MeanSquaredError(), metrics=[keras.metrics.MeanSquaredError()])
```

<h1> Training the Model </h1>

```python
tf.keras.backend.clear_session()  # For easy reset of notebook state.

#creating checkpoints
checkpoint_path = "training1/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

# Create a callback that saves the model's weights
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 save_best_only=True,
                                                 verbose=1)

#train the model

#defining number of epochs and batchsize
epochs = 40
batch_size=64
model.fit(train_generator, steps_per_epoch=len(train_xs)//batch_size,
                    epochs=epochs,
                    validation_data=validation_generator,
                    validation_steps=len(validation_xs)//batch_size,
                    callbacks=cp_callback) # Pass callback to training
```

```python
#validation loss stopped improving after 32nd epoch so we stopped the training
```

```python
#https://www.youtube.com/watch?v=EkzB6PJIcCA&list=PLd9i_xMMzZF6mxod8D1CfqhMy85dyLMq2&index=8

def test_generator(data, batch_size): #the test generator loop does not run forever like data generator
    #runs for one complete iteration
    """
    Yields the next training batch
    
    """
    #data-list of tuples:[(x1,y1),(x2,y2),(x3,y3).....]
    num_samples = len(data)
        
    random.shuffle(data) #shuffling data 
    x,y=zip(*data) #unzipping data to get shuffled x and y
        
    # Get index to start each batch: [0, batch_size, 2*batch_size, ..., max multiple of batch_size <= num_samples]
    for offset in range(0, num_samples, batch_size): 
        # Get the samples you'll use in this batch
        x_batch_sample = x[offset:offset+batch_size]
        y_batch_sample = y[offset:offset+batch_size]

        # Make sure they're numpy arrays (as opposed to lists)-cnn don't take lists as input
        x_batch = np.array(x_batch_sample)
        y_batch = np.array(y_batch_sample)

        # The generator-y part: yield the next training batch            
        yield x_batch,y_batch
```

```python
#creating test generator

batch_size=64

test_generator=test_generator(xy_test,batch_size)
```

```python
model.evaluate(test_generator)
```

<h1> Real Simulation on Test Data </h1>

```python
xs = [] #will store names of all the images
ys = [] #will store the corresponding steering wheel angles

with open("driving_dataset/data.txt") as f:
    for line in f:
        xs.append("driving_dataset/" + line.split()[0])
        #the steering wheel angles are converted from degrees to radians as a method of normalization
        ys.append(float(line.split()[1])) 
```

```python
x_test=xs[34735:]
y_test=ys[34735:]
```

```python
wheel_img = cv2.imread('steering_wheel_image.jpg',0)
rows,cols = wheel_img.shape

smoothed_angle = 0
i=0

while(cv2.waitKey(10) != ord('q')):
    path=x_test[i].split('/')
    path=path[0]+'/images/'+path[1] #defining the image path
    image = cv2.imread(path)
    image_prep=image #preprocessed image
    image_prep=cv2.resize(image,dsize=(200, 66)) #1st arg=new width,2nd arg=new height
    image_prep=image_prep/255.0 #normalizing pixel values between 0 and 1
    image_prep=np.expand_dims(image_prep, 0)  # shape (1, y_pixels, x_pixels, n_bands)-appropriate format for model-4d tensor
    pred_angle=model.predict(image_prep)
    pred_angle=pred_angle*(180/np.pi) #converting to degrees
    pred_angle=pred_angle[0][0] #getting the scalar
    actual_angle=y_test[i]
    print("actual angle: {} predicted angle:{} \n".format(actual_angle,pred_angle))
    cv2.imshow('frame', image)
    smoothed_angle+=0.2*pow(abs((pred_angle-smoothed_angle)),2.0 /3.0)*(pred_angle-smoothed_angle)/abs(pred_angle-smoothed_angle)
    M = cv2.getRotationMatrix2D((cols/2,rows/2),-smoothed_angle,1)
    dst = cv2.warpAffine(wheel_img,M,(cols,rows))
    cv2.imshow("steering wheel", dst)
    i+=1
    if(i==len(x_test)):
        break

cv2.destroyAllWindows()
```
