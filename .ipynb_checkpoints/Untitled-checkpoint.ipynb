<h1> Reading the Dataset </h1>

```python
#the driving dataset contains images in jpg format of various frames captured while driving
#the jpg images are named as- 0.jpg, 1.jpg,.....45405.jpg

#there is a text file named data which has the steering wheel angles in degrees for images from 0.jpg to 45405.jpg

xs = [] #will store names of all the images
ys = [] #will store the corresponding steering wheel angles

with open("driving_dataset/data.txt") as f:
    for line in f:
        xs.append("driving_dataset/" + line.split()[0])
        #the steering wheel angles are converted from degrees to radians as a method of normalization
        ys.append(float(line.split()[1]) * scipy.pi / 180) 
```

```python
#checking if performed above steps properly
print(xs[100:110]) #will print the name of 10 images 
print(ys[100:110]) #will print corresponding steering wheel angles in radians
```

<h1> Splitting the Data- Train, Validation, Test </h1>

```python
#train=80 percent
#validation=20 percent

train_xs = xs[0:int(len(xs) * 0.8)]
train_ys = ys[0:int(len(xs) * 0.8)]

val_xs = xs[int(len(xs) * 0.8):]
val_ys = ys[int(len(xs) * 0.8):]
```

```python
num_train=len(train_xs)
num_val=len(val_xs)

print("number of training images: ",num_train)
print("number of validation images: ",num_val)
```

<h1> Data Preprocessing </h1>

```python
#here the train and validation data is preprocessed to be used by the neural network

#reading an image using opencv
image = cv2.imread('driving_dataset/images/2600.jpg') #--imread() helps in loading an image into jupyter including its pixel values
#displaying the image
#cv2.imshow("sample",image)
# cv2.waitKey(5000) 
# cv2.destroyAllWindows()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
#loading all images and doing the preprocessing

data=[]

for i in tqdm(range(len(xs))):
    path=xs[i].split('/')
    path=path[0]+'/images/'+path[1] #defining the image path
    image = cv2.imread (path)
    image=cv2.resize(image, dsize=(200, 66)) #1st arg=new width,2nd arg=new height
    image=image/255.0 #normalizing pixel values between 0 and 1
    data.append (image)
```

```python
#training data
x_train=data[0:int(len(data) * 0.8)]
y_train=train_ys

#validation data
x_val=data[int(len(data) * 0.8):]
y_val=val_ys

#the above data is too large to be fed as input at once
#so we define data generators which sens in batches and in proper format- note the x are lists of image vectors
#cnn takes numpy arrays only
```

<h1> Defining the Data Generators </h1>

```python
#https://www.youtube.com/watch?v=EkzB6PJIcCA&list=PLd9i_xMMzZF6mxod8D1CfqhMy85dyLMq2&index=8

def data_generator(x, y, batch_size=32):
    """
    Yields the next training batch.
    
    """
    if(len(x)!=len(y)):
        print("lengths of x and y don't match")
        return -1
    
    #x-list of image vectors
    #y-lost of steering angles
    num_samples = len(x)
    while True: # Loop forever so the generator never terminates
        
        # Get index to start each batch: [0, batch_size, 2*batch_size, ..., max multiple of batch_size <= num_samples]
        for offset in range(0, num_samples, batch_size):
            # Get the samples you'll use in this batch
            x_batch_sample = x[offset:offset+batch_size]
            y_batch_sample = y[offset:offset+batch_size]

            # Make sure they're numpy arrays (as opposed to lists)-cnn don't take lists as input
            x_batch = np.array(x_batch_sample)
            y_batch = np.array(y_batch_sample)

            # The generator-y part: yield the next training batch            
            yield x_batch,y_batch
```

```python
#creating train and test generator

batch_size=32

train_generator=data_generator(x_train,y_train,batch_size)

validation_generator=data_generator(x_val,y_val,batch_size)
```

<h1> Defining the CNN Architecture </h1>

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, Dropout
from keras.initializers import TruncatedNormal, Constant


#create model
model = Sequential()
#convolutional layers
model.add(Conv2D(24,kernel_size=5,strides=(2,2),activation='relu', kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1),input_shape=(66,200,3)))
model.add(Conv2D(36, kernel_size=5, strides=(2,2),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
model.add(Conv2D(48, kernel_size=5, strides=(2,2),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
model.add(Conv2D(64, kernel_size=3, strides=(1,1),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
model.add(Conv2D(64, kernel_size=3, strides=(1,1),activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
#flattening layers
model.add(Flatten())
#dense layers with dropout
model.add(Dense(1164,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
model.add(Dropout(0.2))
model.add(Dense(100,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
model.add(Dropout(0.2))
model.add(Dense(50,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
model.add(Dropout(0.2))
model.add(Dense(10,activation='relu',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
model.add(Dropout(0.2))
#final layer with arctan activation as discussed in the paper
model.add(Dense(1, activation='tanh',kernel_initializer=TruncatedNormal(stddev=0.1),
                 bias_initializer=Constant(0.1)))
```

<h1> Summary of the CNN </h1>

```python
model.summary()
```

<h1> Compiling the Model </h1>

```python
#learning rate- 1e-4 (as mentioned in the paper)
#loss- mean square error (regression problem)

model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-4),
              loss=keras.losses.MeanSquaredError(), metrics=[keras.metrics.MeanSquaredError()])
```

<h1> Training the Model </h1>

```python
tf.keras.backend.clear_session()  # For easy reset of notebook state.

#creating checkpoints
checkpoint_path = "training1/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

# Create a callback that saves the model's weights
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 save_best_only=True,
                                                 verbose=1)

#train the model

#defining number of epochs and batchsize
epochs = 30
batch_size=32
model.fit(train_generator, steps_per_epoch=len(x_train)//batch_size,
                    epochs=epochs,
                    validation_data=validation_generator,
                    validation_steps=len(x_val)//batch_size,
                    callbacks=cp_callback) # Pass callback to training
```
